```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
df.info()
```

```python
df['last_four'] = df['CC Number'].apply(lambda x: str(x)[:4])
df.head()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    return '$$$'


df['yelp'] = df['total_bill'].apply(yelp)
df.head()
```
